```python
import matplotlib.pyplot as plt
from pyfht import fht
%matplotlib inline

M = 65536
N = 16384
S = 2000
M_sqrt = 256
sig = 0.2

p_in = np.linspace(1, 5, M)
p_out = np.random.permutation(M)[:N]

def get_x0(M, S):
    """ 真の解 """
    x0 = np.zeros(M)
    x0[np.random.permutation(range(M))[:S]] = np.random.randn(S)
    return x0
    
def forward(x, p_in, p_out, M_sqrt):
    """ システム行列Aの乗算 """
    return fht(p_in * x)[p_out] / M_sqrt

def backward(y, p_in, p_out, M_sqrt):
    """ Aの転置行列の乗算 """
    x = np.zeros(M_sqrt ** 2)
    x[p_out] = y
    return fht(x) / M_sqrt * p_in

def inverse(y, p_in, p_out):
    """ Aの逆行列の乗算 """
    bp = np.zeros(len(p_in))
    bp[p_out] = y
    return fht(bp) / np.sqrt(len(p_in)) / p_in

x0 = get_x0(M, S)
y0 = forward(x0, p_in, p_out, M_sqrt)
b = y0 + np.random.randn(N) * sig

plt.figure(figsize=(10, 7))
plt.plot(x0, lw=0.2)
plt.xlim(0, M)
plt.title(u'真の解')
```

```python
from scipy.optimize import minimize_scalar, minimize

class IterativeShrinkage(object):
    """ 反復縮小アルゴリズム """
    def __init__(self, M, N, p_in, p_out, b, lam, verb=1):
        self.M = M
        self.M_sqrt = np.sqrt(M)
        self.N = N
        self.p_in= p_in
        self.p_out = p_out
        self.b = b
#         self.eps = 1e-10
        self.eps = 0
        self.lam = lam
        self.c = np.max(p_in) ** 2
        self.verb = verb
        
    def forward(self, x):
        """ 順投影 """
        return fht(self.p_in * x)[self.p_out] / self.M_sqrt

    def backward(self, y):
        """ 逆投影 """
        x = np.zeros(self.M)
        x[self.p_out] = y
        return self.p_in * fht(x) / self.M_sqrt

    def shrink(self, x):
        """ ソフト閾値処理 """
        threshold = self.lam / self.c
        return np.where(np.abs(x) > threshold, (np.abs(x) - threshold) * np.sign(x), 0)
    
    def shrink2(self, x, threshold):
        """ ソフト閾値処理2 """
        return np.where(np.abs(x) > threshold, (np.abs(x) - threshold) * np.sign(x), 0)

    def f(self, x, y):
        """ 目標関数 """
        return self.lam * np.sum(np.abs(x)) + 0.5 * np.dot(self.b - y, self.b - y)

    def line(self, x, gx, y, gy):
        """ 直線探索 """
        
        def f_ls(mu, *args):
            """ 目標関数 """
            x, gx, y, gy = args
            x_new = x + mu * gx
            e = self.b - (y + mu * gy)
            return self.lam * np.sum(np.abs(x_new)) + 0.5 * np.dot(e, e)

#         res = minimize_scalar(f_ls, args=(x, gx, y, gy), options={'maxiter': 5})
        res = minimize_scalar(f_ls, args=(x, gx, y, gy))
        return res.x                  
    
    def sesop(self, x, ses_gx, y, ses_gy, q):
        """ 逐次的部分空間最適化 """
    
        def f_sesop(mu, *args):
            """ 目標関数 """
            x, ses_gx, y, ses_gy = args
            x_new = x + np.dot(ses_gx, mu.T)
            e = self.b - (y + np.dot(ses_gy, mu.T))
            return self.lam * np.sum(np.abs(x_new)) + 0.5 * np.dot(e, e)

#         res = minimize(f_sesop, np.zeros(q + 1), args=(x, ses_gx, y, ses_gy), options={'maxiter': 5})
        res = minimize(f_sesop, np.zeros(q + 1), args=(x, ses_gx, y, ses_gy))
        return res.x
    
    def SSF(self, niter=50, ls=False, sesop=False, q=5, x0=None):
        """ 分割可能代理汎関数（separable surrogate functional; SSF） """
        x = np.zeros(self.M)
        y = np.zeros(self.N)
        r = self.b.copy()        
        ses_gx = np.zeros((self.M, q + 1))
        ses_gy = np.zeros((self.N, q + 1))
            
        fs = []
        errs = []
        for k in range(niter):
            fs.append(self.f(x, y))
            if x0 is not None:
                errs.append(np.dot(x - x0, x - x0) / np.dot(x0, x0))
                if k % 20 == 0 and self.verb > 0:
                    print(k, errs[-1])
            e = self.backward(r)
            e_s = self.shrink(x + e / self.c)
            gx = e_s - x
            if sesop:
                gx /= np.sqrt(np.dot(gx, gx))
                ses_gx[:, k % (q + 1)] = gx
                ses_gy[:, k % (q + 1)] = self.forward(gx)
                u = np.dot(ses_gx, self.sesop(x, ses_gx, y, ses_gy, q).T)
            elif ls:
                u = self.line(x, gx, y, self.forward(gx)) * gx
            else:
                u = gx
            x += u
            y = self.forward(x)
            r = self.b - y
            if np.dot(u, u) < self.eps:
                return x, fs, errs
        return x, fs, errs
    
    def IRLS(self, niter=50, ls=False, sesop=False, q=5, x0=None, s_eps=1e-10):
        """ 反復重み付け最小二乗(iterative-reweighted-least-squares; IRLS) """
        x = np.ones(self.M) * 1e-2
        y = self.forward(x)
        r = self.b - y
        ses_gx = np.zeros((self.M, q + 1))
        ses_gy = np.zeros((self.N, q + 1))

        fs = []
        errs = []
        for k in range(niter):
            fs.append(self.f(x, y))
            if x0 is not None:
                errs.append(np.dot(x - x0, x - x0) / np.dot(x0, x0))
                if k % 20 == 0 and self.verb > 0:
                    print(k, errs[-1])
            e = self.backward(r)
            s = (x ** 2) / (2.0 * self.lam / self.c * np.abs(x) + x ** 2 + s_eps)
            e_s = s * (x + e / self.c)
            gx = e_s - x
            if sesop:
                gx /= np.sqrt(np.dot(gx, gx))
                ses_gx[:, k % (q + 1)] = gx
                ses_gy[:, k % (q + 1)] = self.forward(gx)
                u = np.dot(ses_gx, self.sesop(x, ses_gx, y, ses_gy, q).T)
            elif ls:
                u = self.line(x, gx, y, self.forward(gx)) * gx
            else:
                u = gx
            x += u
            y = self.forward(x)
            r = self.b - y
            if np.dot(u, u) < self.eps:
                return x, fs, errs
        return x, fs, errs
    
    def PCD(self, niter=50, ls=False, sesop=False, q=5, x0=None):
        """ 並列座標降下（parallel-coordinate-descent; PCD） """
        x = np.zeros(self.M)
        y = np.zeros(self.N)
        r = self.b.copy()        
        ses_gx = np.zeros((self.M, q + 1))
        ses_gy = np.zeros((self.N, q + 1))
        W = 1. / np.abs(self.backward(self.forward(np.ones(self.M))))
      
        fs = []
        errs = []
        for k in range(niter):
            fs.append(self.f(x, y))
            if x0 is not None:
                errs.append(np.dot(x - x0, x - x0) / np.dot(x0, x0))
                if k % 20 == 0 and self.verb > 0:
                    print(k, errs[-1])
            e = self.backward(r)
            e_s = self.shrink2(x + e * W, self.lam * W)
            gx = e_s - x
            if sesop:
                gx /= np.sqrt(np.dot(gx, gx))
                ses_gx[:, k % (q + 1)] = gx
                ses_gy[:, k % (q + 1)] = self.forward(gx)
                u = np.dot(ses_gx, self.sesop(x, ses_gx, y, ses_gy, q).T)
            elif ls:
                u = self.line(x, gx, y, self.forward(gx)) * gx
            else:
                u = gx
            x += u
            y = self.forward(x)
            r = self.b - y
            if np.dot(u, u) < self.eps:
                return x, fs, errs
        return x, fs, errs
```

```python
niter = 1001
iterative_shrinkage = IterativeShrinkage(M, N, p_in, p_out, b, 1.)
x, fs, errs = iterative_shrinkage.SSF(niter=niter, x0=x0)
x2, fs2, errs2 = iterative_shrinkage.SSF(niter=niter, x0=x0, ls=True)
x3, fs3, errs3 = iterative_shrinkage.SSF(niter=niter, x0=x0, sesop=True)
x4, fs4, errs4 = iterative_shrinkage.IRLS(niter=niter, x0=x0)
x5, fs5, errs5 = iterative_shrinkage.IRLS(niter=niter, x0=x0, ls=True)
x6, fs6, errs6 = iterative_shrinkage.IRLS(niter=niter, x0=x0, sesop=True)
x7, fs7, errs7 = iterative_shrinkage.PCD(niter=niter, x0=x0)
x8, fs8, errs8 = iterative_shrinkage.PCD(niter=niter, x0=x0, ls=True)
x9, fs9, errs9 = iterative_shrinkage.PCD(niter=niter, x0=x0, sesop=True)
```
